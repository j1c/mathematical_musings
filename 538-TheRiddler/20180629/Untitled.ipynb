```python
import random
from itertools import product

import tqdm
```

```python
def get_results(arr):
    nums = range(10)
    all_combinations = list(product(nums, nums, nums, nums))
    all_combinations = [''.join(map(str, x)) for x in all_combinations]
    result = [x in arr for x in all_combinations]
    
    falses = []
    for idx, val in enumerate(result):
        if not val:
            falses.append(all_combinations[idx])
    
    return sum(result), falses
```
